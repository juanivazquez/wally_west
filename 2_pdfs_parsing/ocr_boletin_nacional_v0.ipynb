```python
def get_case(pattern=pattern,doc=doc):
    sp = re.search(pattern, doc).span()
    return doc[:sp[0]],doc[sp[0]:sp[1]],doc[sp[1]:]

def get_case_(pattern=pattern,doc=doc):
    if type(doc)==list:
        for n,d in enumerate(doc):
            try:
                sp = re.search(pattern, d[0]).span()
                doc[n] = [d[:sp[0]],d[sp[0]:]]
            except:
                continue
        return doc
    
    else:
        sp = re.search(pattern, doc).span()
        return [doc[:sp[0]],doc[sp[0]:]]
```
