# Analisis

```python
def check_shares(df: pd.DataFrame, col: str = 'tipo_societario') -> pd.DataFrame:
    # Get value counts
    col_share = df[col].value_counts()

    # Create a DataFrame from the value counts
    df_col = pd.DataFrame(col_share).reset_index()

    # Rename columns for clarity
    df_col.columns = [col, 'count']

    # Calculate total counts
    all_col = df_col['count'].sum()

    # Calculate share and cumulative share
    df_col['share'] = (df_col['count'] / all_col * 100).round(2)
    df_col['share_cum'] = df_col['share'].cumsum()

    return df_col
```

```python
societies_share = df.tipo_societario.value_counts()
df_soc = pd.DataFrame(societies_share)
all_soc = df_soc['count'].sum()
df_soc['share'] = (societies_share/ all_soc * 100).round(2)
df_soc['share_cum'] = df_soc['share'].cumsum()
```
